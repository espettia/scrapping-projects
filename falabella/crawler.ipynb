```python
def extract_category(URL):
    #f = open("web/Televisores-LED grandes ofertas _ falabella.com.html")
    #html_text = f.read()
    webpage = requests.get(URL)
    html_text= webpage.text
    soup = BeautifulSoup(html_text,'html.parser')
    script = soup.find('script',id='__NEXT_DATA__')
    data = js.loads(script.text)
    print("LinkCount:",len(data['props']['pageProps']['results']))
    urls=[]
    for result in data['props']['pageProps']['results']:
        url=result['url']
        urls.append(url)
    return urls
```

```python
def extract_product_info(URL):
    import re
    webpage = requests.get(URL)
    soup = BeautifulSoup(webpage.text,'html.parser')
    product_name = soup.find(class_='fa--product-name').text
    #print(product_name)
    brand_name = soup.find(class_='fa--brand').text
    price_info = soup.find(class_='fa--prices')
    prices = price_info.find_all(string=re.compile('^S/\s*([0-9]+,)*[0-9]+',re.MULTILINE))
    info = {'Nombre':product_name,'Marca':brand_name,'Precios':prices}
    
    specs_main_list = soup.find('div',class_='specifications-list')
    specs_main={}
    if specs_main_list!=None:
        specs_main_html = specs_main_list.find_all('li')
        for spec in specs_main_html:
            key = spec.strong.text
            value = spec.contents[-1]
            specs_main[key]=value
            
    property_name = soup.find_all(class_='property-name')
    property_value = soup.find_all(class_='property-value')
    specs_add_html = zip(property_name,property_value)
    specs_add = {}
    for spec_add_html in specs_add_html:
        specs_add[spec_add_html[0].text]=spec_add_html[1].text
    specs = specs_main | specs_add
    prod = info | specs
    return prod
```

```python
URL='https://www.falabella.com.pe/falabella-pe/product/882877848/Televisor-TCL-Google-TV-55-4K-UHD-55P635'
prod_info = extract_product_info(URL)
```

```python
prods_info=[]
for i in range(1,57):
    URL = "https://www.falabella.com.pe/falabella-pe/category/cat6370551/Televisores-LED?page="+str(i)
    product_pages = extract_category(URL)
    for product_page in product_pages:
        #print(product_page)
        prod_info = {'URL':product_page} | extract_product_info(product_page)
        prods_info.append(prod_info)
```
